```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
f = open("IPO_Data.csv",encoding="utf-8", errors='replace')
data = pd.read_csv(f)
data.head
```

```python
data.shape
```

```python
d = data
index_names = d[(d['Year'] < 2012)].index
d.drop(index_names, inplace = True)
```

```python
d
```

```python
list(data.columns)
```

```python
filtered = d.filter(['ipoDate', 'openDay0', 'Sector', 'DaysBetterThanSP', 'closeDay100'])
filtered.head()
```

```python
filtered.shape
```

```python
filtered.drop(filtered[(filtered['Sector'].isna())].index, inplace = True)
filtered.drop(filtered[(filtered['closeDay100'].isna())].index, inplace = True)
filtered.drop(filtered[(filtered['openDay0'].isna())].index, inplace = True)
```

```python
filtered.shape
```

```python
filtered.head()
```

```python
filtered.rename(columns={'ipoDate': 'date', 'openDay0': 'initialPrice'}, inplace=True)
```

```python
filtered.rename(columns={'Sector': 'sector', 'DaysBetterThanSP': 'daysBeatSP', 'closeDay100': 'closePrice100'}, inplace=True)
```

```python
filtered = filtered.loc[(filtered['initialPrice'] >= 4) & (filtered['initialPrice'] < 1000)]
```

```python
filtered.head()
```

```python
sp_df = pd.read_csv('sp_values.csv')
```

```python
sp_df.head()
```

```python
merged = pd.merge(filtered, sp_df,
                        how="left", on=["date"])
merged.head()
```

```python
merged['priceChange100Days'] = (merged['closePrice100']-merged['initialPrice'])/(merged['initialPrice'])
```

```python
merged = merged.drop(['closePrice100', 'date'], axis=1)
merged.head()
```

```python
merged.shape
```

```python
merged.to_csv('cleaned_ipo_data.csv', index=False)  
```
